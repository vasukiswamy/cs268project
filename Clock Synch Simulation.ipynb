```python
%matplotlib inline
import random
import numpy as np
import matplotlib.pyplot as plt


class Client:

    def __init__(self, id_num, alpha, beta, dist, real_model, GHz):
        self.id = id_num
        self.alpha = alpha
        self.beta = beta
        self.delay = dist / (3.0 * 10**8)
        self.total_error = 0.0
        self.frequency = GHz * 10.0**9  # in Hertz now
        self.real_model = real_model  # if True use the realistic model

        # Variables to be calculated
        # virtual is the first virtual time received - C(t + .... )
        # t0 will be the interval for beacon (sent in first msg)
        self.alpha_est, self.beta_est, self.t0, self.first_clock = (None,) * 4

    # calculates C(t) = a * t + b
    def get_clock_time(self, t):
        if self.real_model:
            n = 0.0
            
            # The correct but really slow way
            # for i in range(int(t * self.frequency)):
            #     n += np.random.normal(0, 0.0005 / (2*self.frequency))  # 0.05%
            # return (self.alpha * t + self.beta + n)

            return self.alpha * t + self.beta + int(t * self.frequency)*np.random.normal(0, 0.0005 / (2*self.frequency))
        else:
            return self.alpha * t + self.beta

    # what the client clock synchs to
    def get_virtual_time(self, t):
        return (self.get_clock_time(t) - self.beta_est) / (self.alpha_est)

    # Receives the first/second synchronization and stores it
    def recv_synch(self, t, t0):
        self.first_clock = self.get_clock_time(t)
        self.t0 = t0

    # Returns the new estimated time
    def update_values(self, iteration, t):
        # Subtract iter by 1 bc the first iter does nothing
        self.alpha_est = (self.get_clock_time(t) - self.first_clock) / (self.t0 * (iteration-1))
        self.beta_est = self.get_clock_time(t) - self.alpha_est * t


class System:

    # t0 is in seconds
    def __init__(self, t0):
        self.num_clients = 0
        self.t0 = t0
        self.current_interval = 0

        # estimated time at time i
        self.estimated_times = {}
        self.real_times = {}

        # key is id; value is client object
        self.clients = {}

    # d is the distance from the master
    def add_client(self, id_num, alpha, beta, dist, real_model, GHz=10):
        self.clients[id_num] = Client(id_num, alpha, beta, dist, real_model, GHz)
        self.estimated_times[id_num] = []
        self.real_times[id_num] = []
        self.num_clients += 1

    # Drifts the clocks a certain number of intervals
    # only call drift once in the beginning for now
    # not sure if calling this later will mess up current_interval for calculations of a, b
    # I think it's ok but not 100% sure
    def drift(self, num_intervals):
        t0 = self.t0
        for i in range(num_intervals):
            for k in range(1, self.num_clients+1):
                # For client k
                # add the client's real time wrt t0* current interval
                self.estimated_times[k].append(self.clients[k].get_clock_time(self.current_interval*t0))
                self.real_times[k].append(self.current_interval*t0)
            self.current_interval += 1

    # starts the synchronization on current interval time
    # only call this once for now
    def start_synch(self, num_times, sigma=0.01, lambda_constant=1, store=False, k_distr=np.random.normal):
        t0 = self.t0
        mean = 0  # this is assumed
        iteration = 1  # stores the interval when updating; ease of use for calculations b/c of drift

        # First time no update
        # Simulate by drifting 1 interval for purposes of graph
        self.drift(1)

        for k in range(1, self.num_clients+1):
            t = self.current_interval * t0 + self.clients[k].delay + lambda_constant * k_distr(mean, sigma)
            self.clients[k].recv_synch(t, t0)

        iteration += 1
        self.current_interval += 1

        # Start updating
        for i in range(num_times-1):
            for k in range(1, self.num_clients+1):
                t = self.current_interval * t0 + self.clients[k].delay + lambda_constant * k_distr(mean, sigma)

                # Update alpha and beta
                self.clients[k].update_values(iteration, t)

                # Calculations for graphs
                real_time = self.current_interval * t0
                est_time = self.clients[k].get_virtual_time(self.current_interval * t0)

                if store:
                    self.estimated_times[k].append(est_time)
                    self.real_times[k].append(real_time)
                else:  # summing for average
                    self.clients[k].total_error += est_time - real_time

            iteration += 1
            self.current_interval += 1

    # Create graph for all time intervals
    def create_graph(self, interval=None):
        if interval:
            intervals = range(interval, len(self.estimated_times[1]))
        else:
            interval = 0
            intervals = range(len(self.estimated_times[1]))

        for i in range(1, 5):
            plt.plot(intervals, abs(np.array(self.estimated_times[i])[interval:] - np.array(self.real_times[i])[interval:]), label='Client ' + str(i))
        plt.legend(loc='upper right')
        plt.ylabel('Difference between estimated and real time')
        plt.xlabel('Time interval')
        plt.show()

    def average_error(self, client_id):
        # interval will always be one more
        # first synch step does nothing/has no error
        return self.clients[client_id].total_error/(self.current_interval - 1)

    # For adding new clocks
    # same as init function
    def reset_system(self):
        self.num_clients = 0
        self.current_interval = 0
        self.estimated_times = {}
        self.real_times = {}
        self.clients = {}

    # time is in seconds
    def create_graph_diff_alpha(self, ppm, step, time, frequency, real_model=False, beta=0.000000003, distance=2.2):
        ppm = ppm/10.0**6
        alpha_values = np.arange(1-ppm, 1+ppm+step, step)  # arange endpoint not inclusive
        num_intervals = int(time/self.t0)  # number of time intervals to run
        results = []

        # Get average longtime error for one alpha value
        for alpha in alpha_values:
            self.reset_system()
            system.add_client(1, alpha, beta, distance, real_model, frequency)
            system.start_synch(num_intervals, sigma=1, lambda_constant=200e-9, store=False)
            results.append(self.average_error(1))
        plt.plot(alpha_values, results)
        plt.ylabel('Average Longtime Error')
        plt.xlabel('Alpha value')
        plt.show()
```

```python
t0 = 0.001 # length of interval; 1 ms
sigma = 1 #for distr of noise

# Initialize
system = System(t0)

# Add some nodes
# id, alpha, beta, distance in meters
system.add_client(1, 0.999999776296498, 0.000000003, 2.2, False)
system.add_client(2, 1.000000141952192, 0.000000001, 2.6, False)
system.add_client(3, 0.999999877467404, 0.000000009, 3.9, False)
system.add_client(4, 0.99999971401255, 0.000000008, 3.1, False)

# Allow clocks to drift 60 intervals
system.drift(60)

# Start synchronization for 50 intervals
system.start_synch(50, sigma, 200e-9, True)

system.create_graph()
system.create_graph(60)
system.create_graph(70)
```
