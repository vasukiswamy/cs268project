```python
%matplotlib inline
import random
import numpy as np
import matplotlib.pyplot as plt
import math


class Client:

    def __init__(self, id_num, alpha, beta, dist, real_model, GHz, t0, duration):
        self.id = id_num
        self.alpha = alpha
        self.beta = beta
        self.delay = dist / (3.0 * 10**8)
        self.total_error = 0.0
        self.frequency = GHz * 10.0**9  # in Hertz now
        self.real_model = real_model  # if True use the realistic model
        self.noise = []

        if real_model:
            prev_sum = 0.0
            sigma = math.sqrt(int(t0 * self.frequency)) * 0.0005 / self.frequency

            # element m in array is the sum of n(1) to n(m)
            for i in range(int(duration/t0)):
                prev_sum += np.random.normal(0, sigma)
                self.noise.append(prev_sum)

        # Variables to be calculated
        # virtual is the first virtual time received - C(t + .... )
        # t0 will be the interval for beacon (sent in first msg)
        self.alpha_est, self.beta_est, self.t0, self.first_clock = (None,) * 4

    # this will save the alpha/beta/noise values
    def reset(self):
        self.total_error = 0.0

    # calculates C(t) = a * t + b
    # kth time interval for real_model
    def get_clock_time(self, t, k=1):
        if self.real_model:
            return (self.alpha * t + self.beta + self.noise[int(k-1)])
        else:
            return self.alpha * t + self.beta

    # what the client clock synchs to
    # kth time interval for real_model
    def get_virtual_time(self, t, k):
        return (self.get_clock_time(t, k) - self.beta_est) / (self.alpha_est)

    # Receives the first/second synchronization and stores it
    def recv_synch(self, t, t0):
        self.first_clock = self.get_clock_time(t, 0)
        self.t0 = t0

    # uses first and last value
    def update_values(self, iteration, t):
        # Subtract iter by 1 bc the first iter does nothing
        self.alpha_est = (self.get_clock_time(t, iteration) - self.first_clock) / (self.t0 * (iteration-1))
        self.beta_est = self.get_clock_time(t, iteration) - self.alpha_est * t

    # uses second to last value
    def update_values_last(self, t, iteration, prev_time):
        self.alpha_est = (self.get_clock_time(t, iteration) - prev_time) / (self.t0)
        self.beta_est = self.get_clock_time(t, iteration) - self.alpha_est * t


class System:

    # t0 is in seconds
    def __init__(self, t0):
        self.num_clients = 0
        self.t0 = t0
        self.current_interval = 0

        # estimated time at time i
        self.estimated_times = {}
        self.real_times = {}

        # key is id; value is client object
        self.clients = {}

    # d is the distance from the master
    def add_client(self, id_num, alpha, beta, dist, real_model, GHz=10, t0=0.001, duration=100):
        self.clients[id_num] = Client(id_num, alpha, beta, dist, real_model, GHz, t0, duration)
        self.estimated_times[id_num] = []
        self.real_times[id_num] = []
        self.num_clients += 1

    # Drifts the clocks a certain number of intervals
    # only call drift once in the beginning for now
    # not sure if calling this later will mess up current_interval for calculations of a, b
    # I think it's ok but not 100% sure
    def drift(self, num_intervals):
        t0 = self.t0
        for i in range(num_intervals):
            for k in range(1, self.num_clients+1):
                # For client k
                # add the client's real time wrt t0* current interval
                self.estimated_times[k].append(self.clients[k].get_clock_time(self.current_interval*t0))
                self.real_times[k].append(self.current_interval*t0)
            self.current_interval += 1

    # starts the synchronization on current interval time
    # only call this once for now
    def start_synch(self, num_times, first=True, sigma=0.01, lambda_constant=1, store=False, k_distr=np.random.normal):
        t0 = self.t0
        mean = 0  # this is assumed
        iteration = 1  # stores the interval when updating; ease of use for calculations b/c of drift

        # First time no update
        # Simulate by drifting 1 interval for purposes of graph
        self.drift(1)

        # used for bad method
        prev_time = {}

        for k in range(1, self.num_clients+1):
            t = self.current_interval * t0 + self.clients[k].delay + lambda_constant * k_distr(mean, sigma)
            self.clients[k].recv_synch(t, t0)
            prev_time[k] = self.clients[k].first_clock

        iteration += 1
        self.current_interval += 1

        # Start updating
        for i in range(num_times-1):
            for k in range(1, self.num_clients+1):
                t = self.current_interval * t0 + self.clients[k].delay + lambda_constant * k_distr(mean, sigma)

                # Update alpha and beta
                if not first:
                    self.clients[k].update_values_last(t, iteration, prev_time[k])
                    prev_time[k] = self.clients[k].get_clock_time(t, iteration)
                else:
                    self.clients[k].update_values(iteration, t)

                # Calculations for graphs
                real_time = self.current_interval * t0
                est_time = self.clients[k].get_virtual_time(self.current_interval * t0, iteration)

                if store:
                    self.estimated_times[k].append(est_time)
                    self.real_times[k].append(real_time)
                else:  # summing for average
                    self.clients[k].total_error += abs(est_time - real_time)

            iteration += 1
            self.current_interval += 1

    # Create graph for all time intervals
    def create_graph(self, interval=None):
        if interval:
            intervals = range(interval, len(self.estimated_times[1]))
        else:
            interval = 0
            intervals = range(len(self.estimated_times[1]))

        for i in range(1, 5):
            plt.plot(intervals, abs(np.array(self.estimated_times[i])[interval:] - np.array(self.real_times[i])[interval:]), label='Client ' + str(i))
        plt.legend(loc='upper right')
        plt.ylabel('Difference between estimated and real time')
        plt.xlabel('Time interval')
        plt.show()

    def average_error(self, client_id):
        # interval will always be one more
        # first synch step does nothing/has no error
        return self.clients[client_id].total_error/(self.current_interval - 1)

    # For adding new clocks
    # same as init function
    def reset_system(self):
        self.num_clients = 0
        self.current_interval = 0
        self.estimated_times = {}
        self.real_times = {}
        self.clients = {}

    # does not delete clocks
    def reset_part_system(self):
        self.current_interval = 0

        for i in self.estimated_times.keys():
            self.estimated_times[i] = []
        for i in self.real_times.keys():
            self.real_times[i] = []

    # time is in seconds
    def create_graph_diff_alpha(self, ppm, step, time, frequency, real_model=False, beta=0.000000003, distance=2.2):
        ppm = ppm/10.0**6
        alpha_values = np.arange(1-ppm, 1+ppm+step, step)  # arange endpoint not inclusive
        num_intervals = int(time/self.t0)  # number of time intervals to run
        results = []

        # Get average longtime error for one alpha value
        for alpha in alpha_values:
            self.reset_system()
            system.add_client(1, alpha, beta, distance, real_model, frequency)
            system.start_synch(num_intervals, first=True, sigma=1, lambda_constant=200e-9, store=False)  # use first clock
            results.append(self.average_error(1))
        plt.stem(alpha_values, results)
        plt.ylabel('Average Longtime Error')
        plt.xlabel('Skew or Alpha Value')
        plt.show()

    # Assumes that the real clock model is always used
    def create_compare_first_consecutive(self, ppm, step, time, frequency, beta=0.000000003, distance=2.2):
        ppm = ppm/10.0**6
        alpha_values = np.arange(1-ppm, 1+ppm+step, step)  # arange endpoint not inclusive
        num_intervals = int(time/self.t0)  # number of time intervals to run
        results_first = []  # use first time
        results_last = []  # use second to last time
        real_model = True  # Always use real clock model

        # Get average longtime error for one alpha value
        for alpha in alpha_values:
            self.reset_system()
            system.add_client(1, alpha, beta, distance, real_model, frequency)

            # Use the first clock time
            system.start_synch(num_intervals, first=True, sigma=1, lambda_constant=200e-9, store=False)  # use first clock
            results_first.append(self.average_error(1))

            # Use the second to last clock time
            self.reset_part_system()
            self.clients[1].reset()

            system.start_synch(num_intervals, first=False, sigma=1, lambda_constant=200e-9, store=False)  # use second to last clock
            results_last.append(self.average_error(1))
        print "Errors with second to last"
        print results_last
        print "Errors with first"
        print results_first
        # plt.plot(alpha_values, results_first)
        # plt.plot(alpha_values, results_last)
        # plt.ylabel('Average Longtime Error')
        # plt.xlabel('Skew or Alpha Value')
        # plt.show()

    def test(self, ppm, step, time, frequency, beta=0.000000003, distance=2.2):
        ppm = ppm/10.0**6
        alpha_values = np.arange(1-ppm, 1+ppm+step, step)  # arange endpoint not inclusive
        num_intervals = int(time/self.t0)  # number of time intervals to run
        results_first = []  # use first time
        results_last = []  # use second to last time
        real_model = True  # Always use real clock model

        self.add_client(1, 0.999999776296498, beta, distance, real_model, frequency)

        # system.start_synch(5, first=True, sigma=1, lambda_constant=200e-9, store=False)  # use second to last clock

        system.start_synch(5, first=False, sigma=1, lambda_constant=200e-9, store=False)  # use second to last clock
        results_last.append(self.average_error(1))
```

```python
# Constants

t0 = 0.001 # length of interval; 1 ms
noise_sigma = 1 # for distr of noise
alpha_ppm = 300
frequency = 10 # Ghz
duration = 100 # seconds
default_model = True # using first clocktime
```
